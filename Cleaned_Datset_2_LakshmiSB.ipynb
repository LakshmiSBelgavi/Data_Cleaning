```python
#Let's check if there are any missing values present in the given dataset.

missing_values = data.isnull().sum()

# Displaying the columns with missing values and their counts.
print("Columns with Missing Values:")
print(missing_values)
```

In the column **"item_name"** lets replace some values as follows :


```python

print(data['item_name'].unique())
```

```python
data['item_name'] = data['item_name'].replace('Chips and Tomatillo-Green Chili Salsa','Chips and Tomatillo Green Chili Salsa')
data['item_name'] = data['item_name'].replace('Chips and Tomatillo-Red Chili Salsa','Chips and Tomatillo Red Chili Salsa')
data['item_name'] = data['item_name'].replace('Chips and Roasted Chili-Corn Salsa','Chips and Roasted Chili Corn Salsa')
```

The **"choice_description"** column : replace the NaN values to None and removing [  ] brackets.

```python
print(data['choice_description'].unique())
```

```python

data['choice_description'] = data['choice_description'].apply(lambda x: None if pd.isna(x) else x.strip('[]'))
```

```python
data['choice_description'] = data['choice_description'].str.replace("[","")
data['choice_description'] = data['choice_description'].str.replace("]","")
```

```python
print(data['choice_description'].unique())
```

```python

df_choices = data['choice_description'].str.split(',', expand=True)

# Renaming the new columns to 'choice1', 'choice2', etc.
df_choices.columns = [f'choice{i+1}' for i in range(df_choices.shape[1])]

# Concatenating the new columns with the original DataFrame.
data = pd.concat([data, df_choices], axis=1)

```

```python
data
```

```python
# Drop the original 'choice_description' column if needed
data.drop('choice_description', axis=1, inplace=True)
```

In the **"item_price"** column : replace the $ symbol and convert the datatype of that column to int float datatype.

```python
data['item_price'] = data['item_price'].str.replace("$","")
```

```python
data['item_price']
```

```python
# Converting the 'item_price' column to a numeric data type
data['item_price'] = pd.to_numeric(data['item_price'])
```

```python
data.info()
```

```python
#Lets drop the duplicate values.
data.drop_duplicates(keep=False, inplace=True)
```

```python
data.shape
```

```python
#Re-arranging the columns as required
column_order = ['order_id','quantity','item_name','choice1','choice2','choice3','choice4','choice5','choice6','choice7','choice8','choice9','choice10','item_price']
data = data[column_order]
```

# The final **Cleaned Dataset** is as follows **:)**

```python
data
```

## **1) Get the quantity of each item.**

```python

total_quantity = data.groupby('item_name')['quantity'].sum().reset_index()

print(total_quantity)
```

## **2)What was the most ordered item in the choice_description column?**

```python

choices_columns = data[['choice1', 'choice2', 'choice3', 'choice4', 'choice5', 'choice6', 'choice7', 'choice8', 'choice9', 'choice10']]


all_choices = choices_columns.values.flatten()

# Filtering out 'None' values
non_none_choices = all_choices[~pd.isna(all_choices) & (all_choices != 'None')]

# Finding the most ordered item
most_ordered_item = pd.Series(non_none_choices).value_counts().idxmax()

print("The most ordered item is:", most_ordered_item)

```

## **3)How many items were ordered in total?**

```python

total_items_ordered = data.groupby('item_name')['quantity'].sum().reset_index()

total_items = total_items_ordered['quantity'].sum()

print("Total number of items ordered:", total_items)
```

## **4)How much was the revenue for the period in the dataset?**


```python

data['total_price'] = (data['quantity']) * (data['item_price'])

total_revenue = data['total_price'].sum()

print("Total revenue for the period:", total_revenue)
```

## **5)How many orders were made in the period?**


```python

total_orders = data['order_id'].nunique()

print("Total number of orders made in the period:", total_orders)
```

## **6)What is the average revenue amount per order?**

```python

average_revenue_per_order = total_revenue / total_orders

print("Average revenue amount per order:", average_revenue_per_order)
```

## **7)How many different items are sold?**

```python

num_different_items = data['item_name'].nunique()

print("Number of different items sold:", num_different_items)
```
